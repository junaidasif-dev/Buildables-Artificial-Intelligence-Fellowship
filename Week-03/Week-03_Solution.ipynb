# **Questions From "1 Start Writing Prompts Like a Pro"**

 
1. **What is prompting?**  
‚û°Ô∏è Prompting is the process of giving clear and specific instructions or inputs to a generative AI tool so it produces the desired output.

2. **In the following prompt, which part describes the primary task?**  
*Prompt: ‚ÄúExplain how car batteries work using non-technical language that someone without any mechanical knowledge could understand and interpret.‚Äù*  
‚û°Ô∏è The primary task is: ‚ÄúExplain how car batteries work.‚Äù

3. **When you are using a gen AI tool to create a healthy three-course menu that requires a certain number of calories, which prompting element will ensure the menu is more practical?**  
‚û°Ô∏è Adding constraints (specific calorie requirement) makes the menu more practical.

4. **Prompting with audio to receive a visual representation is called what?**  
‚û°Ô∏è That‚Äôs called multimodal prompting.

5. **What does a human-in-the-loop approach to prompting emphasize?**  
‚û°Ô∏è It emphasizes human oversight, feedback, and iterative refinement to guide the AI toward accurate and useful results.

6. **When the initial prompt "Create a dataset" is too broad, what should you do to refine it?**  
‚û°Ô∏è You should add context, details, and constraints (e.g., type of dataset, size, format, domain) to make the request more specific and actionable.

# **Questions from "3 Use AI as a creative or expert partner"**

### üìù Questions for Reflection

1. How is prompt chaining similar to solving a jigsaw puzzle?  
Because both involve breaking a big, complex problem into smaller, manageable pieces and then combining them to form the complete picture. Each ‚Äúprompt‚Äù is like a puzzle piece that contributes to the final solution.  

2. What‚Äôs the main difference between iteration and chaining?  
- Iteration = refining the same prompt multiple times to improve the output.
- Chaining = breaking a complex task into a sequence of smaller prompts, where each output feeds into the next step.

3. Why might prompt chaining be more effective than asking AI one big question?  
Because one big question can overwhelm the model, leading to vague or unfocused results. Chaining keeps the task structured, guides the model step by step, and ensures higher accuracy and clarity.

4. How can the context window of an AI model affect your chaining strategy?  
The context window limits how much information the model can ‚Äúremember‚Äù at once. If your chain is too long, earlier details may get lost. To manage this, you might need to:
    - Summarize intermediate outputs.
    - Use external storage (notes, databases) to carry context forward.
    - Keep each step concise.

### ‚úÖ Reflection Question üí°
How could CoT and ToT prompting help you **solve complex problems** or **spark creativity** in your workplace?

- **Decompose complexity:** CoT (stepwise reasoning) breaks decisions into traceable steps so you can audit and trust the result.  
- **Explore more possibilities:** ToT (branching / multiple-thought paths) forces the model to propose diverse options rather than one shallow idea.  
- **Better trade-offs:** Combining them lets you generate many options (ToT) and then rigorously justify and refine one (CoT), improving quality of final plans.  
- **Faster iteration:** Clear, numbered steps and evaluation metrics make it easy to tweak assumptions and re-run specific parts of the chain.  
- **Human + AI collaboration:** Structured outputs (scores, short justifications, step plans) make it easy for humans to review, challenge, and approve ‚Äî ideal for human-in-the-loop workflows.
